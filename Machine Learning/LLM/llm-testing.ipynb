## Model Tools

```python
import json
from time import time
import numpy as np
# from huggingface_hub import InferenceClient
import voyageai

# Separate cohere client used only for embeddings
vo = voyageai.Client(api_key=VOYAGE_API_KEY)
# EMBED_MODEL = "sentence-transformers/all-MiniLM-L6-v2"
BATCH_SIZE = 32

# Parse D3FEND/MITRE JSON-LD KB
with open("knowledgebase.json", "r") as f:
    kb_raw = json.load(f)

graph = kb_raw.get("@graph", [])

def extract_str(val) -> str:
    """Handle JSON-LD values that may be a plain string, a dict, or a list."""
    if isinstance(val, str):
        return val
    if isinstance(val, dict):
        return val.get("@value", "")
    if isinstance(val, list) and val:
        return extract_str(val[0])
    return ""

# Extract entries that have both a label and a definition
kb_entries = []
for node in graph:
    title   = extract_str(node.get("rdfs:label", ""))
    content = extract_str(node.get("d3f:definition", ""))
    if title and content:
        kb_entries.append({
            "id":      node.get("@id", ""),
            "title":   title,
            "content": content,
        })

kb_texts = [e["title"] + " " + e["content"] for e in kb_entries]

# Embed in batches via Groq API (no torch needed)
all_vecs = []
for i in range(0, len(kb_texts), BATCH_SIZE):
    batch = kb_texts[i : i + BATCH_SIZE]
    vecs = np.array(vo.embed(batch, model="voyage-3-lite", input_type="document").embeddings)

    all_vecs.append(vecs)
    if (i // BATCH_SIZE) % 10 == 0:
        print(f"  Embedded {i + len(batch)}/{len(kb_texts)} entries...")

kb_vectors = np.vstack(all_vecs)
# Normalise rows for cosine similarity (dot product == cosine when normalised)
kb_vectors = kb_vectors / np.linalg.norm(kb_vectors, axis=1, keepdims=True)

print(f"KB index built: {len(kb_entries)} entries, dim={kb_vectors.shape[1]}")
```

```python
import requests
from requests.auth import HTTPBasicAuth

def create_jira_ticket(summary: str, description: str, priority: str = "Medium") -> dict:
    """Create a JIRA ticket and return the issue key and URL."""
    url = JIRA_REQUEST_URL
    auth = HTTPBasicAuth(JIRA_AUTH_EMAIL, JIRA_API_TOKEN)
    headers = {
        "Accept": "application/json",
        "Content-Type": "application/json"
    }
    payload = {
        "fields": {
            "project":     {"key": JIRA_PROJECT_KEY},
            "summary":     summary,
            "description": {
                "type": "doc",
                "version": 1,
                "content": [{"type": "paragraph", "content": [{"type": "text", "text": description}]}]
            },
            "issuetype":   {"name": "Bug"},
            "priority":    {"name": priority},
            "labels":      ["honeypot", "automated"],
        }
    }

    try:
        response = requests.post(url, json=payload, headers=headers, auth=auth)
        response.raise_for_status()
        key = response.json().get("key")
        print(f"✓ Jira ticket created: {key}")
        return {"key": key}
    except requests.exceptions.RequestException as e:
        print(f"✗ Failed to create Jira ticket: {e}")
        return {}
```

```python
def query_knowledgebase(query: str, top_k: int = 3) -> str:
    """Semantic search over the D3FEND/MITRE KB — called by the model as a tool."""
    # q_vec = np.array(client.feature_extraction(query, model=EMBED_MODEL))
    q_vec = np.array(vo.embed([query], model="voyage-3-lite", input_type="document").embeddings)[0]
    q_vec = q_vec / np.linalg.norm(q_vec)

    scores  = kb_vectors @ q_vec
    top_idx = np.argsort(scores)[::-1][:top_k]

    results = []
    for i in top_idx:
        if i < len(kb_entries):
            e = kb_entries[i]
            results.append(
                f"[KB-ID: {e['id']}] {e['title']}\n"
                f"Content: {e['content']}"
            )
    return "\n\n---\n\n".join(results)
```

```python
import json
import os
# from huggingface_hub import InferenceClient
from groq import Groq

client = Groq(api_key = os.getenv("GROQ_API_KEY"))

messages = [
    {"role": "system", "content": f"HISTORICAL_CLUSTER_CONTEXT:\n{HDBSCAN_clusters_after_training}"},
    {"role": "system", "content": f"CURRENT_ROLLUP:\n{rollup_payload1}"},
    {"role": "system", "content": f"CURRENT_PREDICTIONS:\n{pred}"},
    {"role": "system", "content": f"CLUSTER-SEVERITY-LEVELS:\n{cluster_severity_dict}"},
    {"role": "system", "content": f"CURRENT_CLOUDFLARE_WAF_RULES:\n{current_WAF_rules}"},
    {"role": "user",   "content": prompt},
]

tools = [
    {
        "type": "function",
        "function": {
            "name": "query_knowledgebase",
            "description": (
                "Semantic search over the internal cybersecurity knowledgebase. "
                "Use this to look up CVEs, MITRE ATT&CK tactics, remediation guides, "
                "or threat intel relevant to a detected cluster or pattern."
            ),
            "parameters": {
                "type": "object",
                "properties": {
                    "query": {
                        "type": "string",
                        "description": "Natural language query e.g. 'PHP webshell CVE remediation' or 'MITRE T1190'"
                    },
                    "top_k": {
                        "type": "integer",
                        "description": "Number of KB entries to return (default 3)",
                        "default": 3
                    }
                },
                "required": ["query"]
            }
        }
    },
    {
        "type": "function",
        "function": {
            "name": "create_jira_ticket",
            "description": "Create a JIRA security ticket for a detected attack cluster.",
            "parameters": {
                "type": "object",
                "properties": {
                    "summary":     {"type": "string"},
                    "description": {"type": "string"},
                    "priority":    {"type": "string", "enum": ["Highest", "High", "Medium", "Low"]}
                },
                "required": ["summary", "description"]
            }
        }
    }
]

# --- Tool dispatch ---
def dispatch_tool(call):
    args = json.loads(call.function.arguments)
    if call.function.name == "query_knowledgebase":
        return query_knowledgebase(**args)
    elif call.function.name == "create_jira_ticket":
        args.setdefault("priority", "Medium")
        return json.dumps(create_jira_ticket(**args))
    return json.dumps({"error": "unknown tool"})

# Tool dispatch loop — keeps going until the model stops calling tools
MAX_ITERS = 10
force_finish = False

for _ in range(MAX_ITERS):
    completion = client.chat.completions.create(
        model="llama-3.3-70b-versatile",
        messages=messages,
        tools=tools,
        tool_choice="none" if force_finish else "auto",
        stream=False
    )

    msg = completion.choices[0].message

    if msg.tool_calls and not force_finish:
        messages.append({"role": "assistant", "tool_calls": msg.tool_calls})
        for call in msg.tool_calls:
            result = dispatch_tool(call)
            messages.append({
                "role": "tool",
                "tool_call_id": call.id,
                "content": json.dumps(result),
            })
            if call.function.name == "create_jira_ticket":
                force_finish = True  # next iteration must give final response
    else:
        print(msg.content)
        break
```
